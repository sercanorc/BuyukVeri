2-) Lisans not ortalaması 10 üzerinde 8'in altında olan kişi sayısı

```python
query = "SELECT count(*) as KisiSayisi from Admission_Predict_Ver1 where CGPA<8"
df = pd.read_sql_query(query,connection)
df
```

3-) Yüksek lisans başvurus yapan kişilerin ortalama TOEFL Scoreları

```python
query = "SELECT AVG(TOEFLScore) as OrtalamaTOEFLScore from Admission_Predict_Ver1"
df = pd.read_sql_query(query,connection)
df
```

4-)TOEFL Score'u 100 ün altında olan kişilerin ortalama yüksek lisansa kabul şansı

```python
query = "SELECT AVG(ChanceofAdmit) as KabulSansı from Admission_Predict_Ver1 where TOEFLScore<100 "
df = pd.read_sql_query(query,connection)
df
```

5-) Üniversite puanına göre kabul şansı gruplaması

```python
query = "SELECT UniversityRating, AVG(ChanceofAdmit) as KabulSansı from Admission_Predict_Ver1 GROUP BY UniversityRating"
df = pd.read_sql_query(query,connection)
df
```

6-)En düşük Gre puanı ve en yüksek Gre Puanı

```python
query = "SELECT MIN(GREScore)as MİN, MAX(GREScore)as MAX from Admission_Predict_Ver1 ORDER BY GREScore"
df = pd.read_sql_query(query,connection)
df
```

7-)Yüksek lisans kabul şansı %80den fazla olan kişilerin lisans not ortalaması en düşük 10 kişi

```python
query = "SELECT * from Admission_Predict_Ver1 where ChanceofAdmit>0.9 ORDER BY CGPA LIMIT 10"
df = pd.read_sql_query(query,connection)
df
```

```python
8-) Lisans not ortalaması 7.5 den küçük araştırma deneyimi olan kişi
```

```python
query = """SELECT * from Admission_Predict_Ver1 where CGPA<7.5 AND Research=1 ORDER BY ChanceofAdmit DESC """
df = pd.read_sql_query(query,connection)
df
```

9-)Kabul şansı 0.8 ile 1 arasında ve araştırma deneyimi olan kişilerin GREScore TOEFLScore ve Lisans not ortalamaları	

```python
query = """SELECT GREScore,TOEFLScore,CGPA from Admission_Predict_Ver1 
where 
ChanceofAdmit IN (0.8,1) AND Research=1 
ORDER BY ChanceofAdmit DESC """
df = pd.read_sql_query(query,connection)
df
```

10-)Kişilerin GREScore TOEFLScore ve Lisans not ortalamaları araştırma deneyimine göre dağılımı

```python
query = """SELECT Research, AVG(GREScore),AVG(TOEFLScore),AVG(CGPA) from Admission_Predict_Ver1  
GROUP BY Research """
df = pd.read_sql_query(query,connection)
df
```

11-)Kabul şansında Amaç Beyanı ve Tavsiye Gücü Mektubunun dağılımı

```python
query = """SELECT SOP, AVG(ChanceofAdmit) AS KabulŞansı
from Admission_Predict_Ver1  
GROUP BY SOP """
df = pd.read_sql_query(query,connection)
df
```

12-)Kabul şansında üniversite puanının dağılımı

```python
query = """SELECT UniversityRating, AVG(ChanceofAdmit) AS KabulŞansı
from Admission_Predict_Ver1  
GROUP BY UniversityRating """
df = pd.read_sql_query(query,connection)
df
```

```python
query = """SELECT DISTINCT UniversityRating, AVG(SOP) AS TavsiyeMektubu 
FROM 
Admission_Predict_Ver1 
GROUP BY UniversityRating"""
df = pd.read_sql_query(query,connection)
df
```
